- Phonenumbers are strings and are that way because we need string methods rather than numerical methods when using the numbers eg concat(), len() etc but you'll never need the numeric methods in phonenumbers eg the addition or subtraction. Here we want to verify if our phone number has indeed 12 characters as it should using the `LENGTH()` METHOD.

```python
%%sql
SELECT
    LENGTH(phone_number)
from
    employee
LIMIT
    5;
```

- We see that the length is 13 rather than 12
- 

```python
%%sql
SELECT
    TRIM(phone_number) AS new_len
from
    employee;
```

```python
%%sql
UPDATE
    employee
SET
    phone_number=TRIM(phone_number);
```

- We have successfully updated the phone_numbers column using the `TRIM()` STRING 'METHOD'

## Counting the number of residents in the location. We use the window function to allow us to the distinct rows and then group them over the unique...

```python
%%sql
SELECT
    DISTINCT town_name,
    COUNT(town_name) OVER(PARTITION BY town_name
                         ORDER BY town_name) AS number_of_residents
FROM
    employee;
```
